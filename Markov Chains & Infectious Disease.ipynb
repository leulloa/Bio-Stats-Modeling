# Markov Chains & Infectious Disease

**1** <span style="color:blue"> The probability of a healthy person
staying healthy the ned day is 0.85 since the arrows leaving the state
of healthy must sum to 1. In other words, 1 - 0.15 = 0.85. The
probability of an infected person still being infected the next day is
0.80. Alternatively, 1 - 0.20 = 0.80. </span>

**2** <span style="color:blue"> The code is simulating the spread of an
infectious disease over 100 days. The first line initializes a variable
‘days’ to the value 100, which represents the number of days in the
simulation.</span>

<span style="color:blue"> The second line initializes a variable
‘CURstate’ to 0. In this case, 0 represents the healthy state, while 1
represents the infected state.By initializing “CURstate’ to 0, we are
starting with the assumption that the individual is initially healthy.
This is necessary to track the progression of the individual’s health
state over time, starting from an initial known state.</span>

<span style="color:blue"> The third line creates an empty vector of
length ‘days’ called ‘state’ to store the state of the population for
each day of the simulation.</span>

<span style="color:blue"> ‘for(i in 1:days)’ initiates a for loop that
will iterate ‘days’ times, with the index variable ‘i’ taking on values
1 through ‘days’. </span>

<span style="color:blue"> ‘randomNum \<- runif(1)’ generates a random
number between 0 and 1, and assigns it to the variable ‘randomNum’.
</span>

<span style="color:blue"> The ‘if’ statements use the current state of
the population (‘CURstate’) and the value of ‘randomNum’ to determine
the infection state of the population at the next time step. </span>

<span style="color:blue"> If ‘CURstate’ is 0 (representing a healthy
state), and ‘randomNum’ is less than 0.15, then ‘CURstate’ is changed to
1 (infected state). Otherwise, ‘CURstate’ remains 0. </span>

<span style="color:blue"> If ‘CURstate’ is 1 (representing a infected
state), and ‘randomNum’ is less than 0.20, then ‘CURstate’ is changed to
0 (healthy state). Otherwise, ‘CURstate’ remains 1. </span>

<span style="color:blue"> ‘state\[i\] \<- CURstate’ assigns the current
value of ‘CURstate’ to the ith element of the ‘state’ vector,
representing the infection state of the population at time i. </span>

<span style="color:blue"> Lastly, the ‘state’ vector is returned as the
output of the code. </span>

<span style="color:blue"> The purpose of ‘runif(1)’ in the 5th line is
to generate a single random number between 0 and 1. This random number
is used to determine whether an individual transitions from a healthy
state to an infected state on a given day or transitions from an
infected state to a healthy state.The purpose of using a random number
in this simulation is to introduce some stochasticity or randomness into
the model. This randomness is meant to represent the unpredictable
nature of disease transmission and the varying likelihood of a person
becoming infected on any given day. </span>

<span style="color:blue"> If ‘randomNum \< .15’, the the individual
transitions from a healthy state (‘CURstate = 0’) to an infected state
(‘CURstate = 1’) on that day. The threshold value of .15 represents the
probability that an individual will transition from healthy and become
infected on a given day. This value can be thought of as the
transmission rate of the disease. If the random number generated by
‘runif(1)’ is \< .15, then transmission occurs, and the individual’s
state changes to infected. If the random number is greater than or equal
to .15, then the individual ramins healthy.</span>


<span style="color:blue"> The plot shows an initial rapid increase in
the number infected individuals (opposite the previous plot). Since the
initial states of all 10,000 individuals was changed to healthy, the
first data point occurs where both the number of days and the number of
infected individuals is 0. As the number of days increases, the number
of infected individuals fluctuates. This is because the disease may
spread quickly in the beginning, when there are many susceptible
individuals, but as more individuals become infected, there are fewer
susceptible individuals remaining and the rate of spread slows down. In
addition, the fluctuations in the number of infected individuals over
time may be due to random chance events or changes in the behavior of
the population. </span>

# GILLESPIE ALGORITHM SIMULATOR FOR USE IN 24 AND 25 - NO NEED TO COPY-PASTE

gillespie <- function(counts, t_max, contact, recovery, birth){
  # gillespie() takes in the variables explained below and runs gillespie simulation of a Continuous-Time Markov Chain
  
  # counts: The initial number of people within each group (SIR)
  # t_max: The maximum time we want to simulate to
  # contact: Rate at which people come into contact
  # recovery: Rate at which people recover from illness
  # birth: Rate at which new births enter the population (equivalent to people dying too)

  # Should only run if we have at least 1 infection
  if (counts[2] == 0) {
    # No infections in this population, no point of running
    stop("Must initialize with at least one infected person in I (compartment 2)")
    geterrmessage()
    return()
  }

  # Initial time
  t <- 0.0
  
  # Data matrix storing time and counts within each compartment
  data_mat <- matrix(0, nrow = 1, ncol = 4)
  data_mat[1, ] <- c(t, counts)
  
  while (t < t_max) {
    # SIR case, only care about infected in compartment 3 
    if (counts[2] == 0) {
        break
    }

    # Calculate propensities of each reaction (S -> I, I -> R, I -> S, R -> S)
    p <- numeric(4)
    # S -> I Reaction Propensity: (Number Susceptible)(Number Infected)(Contact Rate)/(Total People)
    p[1] <- counts[1] * counts[2] * contact / sum(counts)
    # I -> R Reaction Propensity: (Number Infected)(Recovery Rate)
    p[2] <- counts[2] * recovery
    # I -> S Reaction Propensity: (Number Infected)(Birth/Death Rate)
    p[3] <- counts[2] * birth
    # R -> S Reaction Propensity: (Number Recovered)(Birth/Death Rate)
    p[4] <- counts[3] * birth
    
    # Generate next time step
    dt <- rexp(1, sum(p))
    t <- t + dt
    
    # Generate next reaction and update counts
    reac <- runif(1)
    if (reac < p[1]/sum(p)) {
      counts[1] <- counts[1] - 1
      counts[2] <- counts[2] + 1
    } else if (reac < sum(p[1:2])/sum(p)) {
      counts[2] <- counts[2] - 1
      counts[3] <- counts[3] + 1
    } else if (reac < sum(p[1:3])/sum(p)) {
      counts[2] <- counts[2] - 1
      counts[1] <- counts[1] + 1
    } else {
      counts[3] <- counts[3] - 1
      counts[1] <- counts[1] + 1
    }
    
    # Add time and counts to vector of vectors data
    data_mat %<>% rbind(c(t, counts))
  }
  
  # Format data in tibble
  colnames(data_mat) <- c("Time", "S", "I", "R")
  data_tib <- as_tibble(data_mat)
  
  # Return as nice ggplot
  ggplot(data = data_tib) + geom_path(mapping = aes(x = Time, y = S, color = 'a')) +
    geom_path(mapping = aes(x = Time, y = I, color = 'b')) +
    geom_path(mapping = aes(x = Time, y = R, color = 'c')) +
    scale_color_manual(guide = 'legend', values = c('a' = 'red', 'b' = 'blue', 'c' = 'black'), labels = c('S', 'I', 'R')) +
    labs(x = "Time (arbitrary units)", y = "Number of People", title = "Simple SIR Test", colour = "Compartment")
}
