**1.2** <span style="color:blue"> We test each sample multiple times to
reduce the propensity for a false positive/negative result and to
increase the accuracy and reliability of the overall result. ELISA tests
are not always 100% accurage, and there may be factors that affect the
test results (sample handling, human error, issues with the test kit
itself, etc.). By testing each sample multiple times, we may be able to
obtain a more robust and accurate assessment of the patient’s HIV

**1.3** <span style="color:blue"> In an ELISA test, the patient’s sample
changes color if it contains antibodies against the HIV virus. The color
change occurs as a result of an enzymatic reaction that takes place on
the surface of the ELISA plate, and the intensity of the color may be
proportional to the amount of antibodies present in the sample. If a
patient’s sample changes color, it indicates that they have likely been
exposed to the HIV virus and their body has had an immune response
against it. A positive sample may begin to resemble the positive control
in color. </span>

**1.4** <span style="color:blue"> It is important to use two different
antibodies since they each fulfill a speicific purpose. The first
antibody is immobilized on the surface of the ELISA plate and used to
capture the HIV antigen from the patient sample (if the antigen is
present). The secondary antibody (an antibody that recognizes human
antibodies) is labeled with an enzyme and is used to detect the captured
antigen. After the capture antibody captures the antigen, the detection
antibody binds to the captured antigen forming a complex (called
conjugate). The enzyme linked to the detection antibody converts the
colorless substrate into a colored product to indciate the presence of
the antigen (if the antigen is present). </span>

**1.5** <span style="color:blue"> Each step in the ELISA test serves a
specific purpose and builds on the previous step. As a result, it is
important that the steps are conducted in order. If they are not, a
false positive or negative may result. The steps for an ELISA test are
as follows: \[1\] Antigen proteins added to wells and bind to the bottom
\[2\] Blood serum for a patient is added (could contain primary
antibodies) \[3\] Wells are washed with a buffer to remove unbound
material (skipped in this lab) \[4\] Secondary antibodies are added
\[5\] Detection process \[6\] Washing (skipped in this lab). If the
order of these steps is changed, as an example, the second antibody may
bind to a non-specific protein on the plate instead of the target
antigen. This may lead to a flase positive result. </span>

**1.6** <span style="color:blue"> The blocking buffer is important in an
ELISA test as it can help to prevent non-specific binding of the
secondary antibody and other unwanted proteins to the surface of the
well or the membrane. The blocking buffer coats the well surface and
fills in any remaining unoccupied binding sites on the sufrace that have
the potential to ineract with the secondary antibody or other proteins
in the sample. This may reduce the background signal and help to
increase the sensitivity and specificity of the assay. In other words,
it could help to minimize false positives/negatives and to ensure the
accuracy and reliability of the results. </span>

**1.7** <span style="color:blue"> Mixing the contents of each well with
the addition of every reagent in an ELISA test helps to ensure that each
well is uniformly exposed to the reagants and that the reactions can
occur consistently across all wells. Since the ELISA test relies on the
specific binding of antibodies to antigens, any variation in the
distribution of reagents across wells could lead to inaccurate results.
Mixing the contents helps to minimize experimental variability and to
ensure the reliability of the test results. </span>

**1.8** <span style="color:blue"> An ELISA experiment for COVID-19 might
resemble the following: </span>

<span style="color:blue"> 1. Identify specific antigens or antibodies

<span style="color:blue"> 3. Block the unoccupied sites in the wells

<span style="color:blue"> 5. Add the patient samples to the wells and

</span>

<span style="color:blue"> 6. Wash the plate to remove any unbound

<span style="color:blue"> 7. Add a secondary antibody specific to the

</span>

<span style="color:blue"> 8. Wash the plate again to remove any unbound

</span>

<span style="color:blue"> 9. Add a substrate for the enzyme label.
Substrate may be a chromogenic/fluorogenic molecule. </span>

<span style="color:blue"> 10. Measure the signal from each well using a

Signal from each well should indicate the presence or absence of
specific antibodies or antigens present in the patient sample. </span>

<span style="color:blue"> TO ENSURE THE ELISA TEST IS RELIABLE: Include


    ## [1] 12.36241

<span style="color:blue"> In question 2.1, the funkySampler function was
used to generate multiple samples of fixed size and number of trials,
whereas in question 2.3, the sample function was used to generate a
single sample of varying sizes. Generally, as sample size increases,
standard error of the mean decreasses regardless of which functionw as
used. Larger samples provide more information about the underlying
population and should therefore reduce the amount of random error in the
estimate of the mean. Increasing the number of trials may some (but
limited) effect on the standard error of the mean (it is not included
directly in the calculation, but it is included in the generation of a
random sample by the funkySampler function), particularly if the sample
size is large enough to provide a reliable estimate of the mean.
Ideally, the standard error of the mean should be highest when the
sample size is lowest. It should decrease from a sample size of 7 to 8
to 9 to 17, 19, 170, 199, etc. However, the samples are random and the
functions used to calculate the standard error of the mean are different
(this may not have a significant effect since both generate random
samples), so this may not be the case in practice.</span>

**3.1 (a)** <span style="color:blue"> We used qnorm(0.975) instead of
qnorm(0.95) to find the 95% confidence interval because the normal
distribution is symmetric around the mean, so the area under the curve
to the left of the mean is the same as the area to the right of the
mean. Therefore, we must find the point on the distribution that leaves
2.5% in the lower tail and 2.5% in the upper tail, which is equivalent
to finding the point that leaves 95% in the middle. This point is in the
97.5th percentile of the standard normal distribution. Hence, we use
qnorm(0.975) to find the z-score for the upper end of the 95% confidence
interval. </span>

**3.1 (b)** <span style="color:blue"> In other words, this question is
asking: how might we find the 95% confidence interval given the z-score
corresponding to the 97.5th percentile of the standard normal
distribution. The margin of error can be calculated by multiplying the
z-score by the standard error of the mean (which is equal to the sample
standard deviation divided by the square root of the sample size).
Lastly, the lower and upper bounds of the confidence interval can be
calculated by subratcing and adding the margin of error to the sample
mean, respectively. The bounds can be rounded to two decimal places and
combined into a string. The resulting interval should contain the true
populate mean with a 95% probability. </span>

**3.1 (c)** <span style="color:blue"> Begin by creating variables for
the sample mean, standard deviation, and sample size. Then, calculate
the 98% confidence interval using the qnorm function with a probability
value of 0.99 (since the qnorm function gives the z-value for the upper
tail of the distribution, we should use 0.99 instead of 0.98 to obtain
the correct value for a 98% confidence interval). Calculate the lower
and upper bounds of the confidence interval by subtracting and adding
the margin of error, respectively. Lastly, round the bounds to two
decimal places and combine them into a string. An example of this code
might look as follows: </span>

<span style="color:blue"> sampleMean \<- 10 </span>

<span style="color:blue"> SD \<- 3 </span>

<span style="color:blue"> n \<- 30 </span>

<span style="color:blue"> error \<- qnorm(0.99) \* SD / sqrt(n) </span>

<span style="color:blue"> lower \<- round(sampleMean - error, digits =
2) </span>

<span style="color:blue"> upper \<- round(sampleMean + error, digits =
2) </span>

<span style="color:blue"> CI98 \<- paste(toString(lower), “,”,
toString(upper)) </span>

<span style="color:blue"> CI98 </span>

**3.2**

``` r
CI95 <- function(x){
  n <- length(x)
  sampleMean <- mean(x)
  SD <- sd(x)
  error <- qnorm(0.975) * SD / sqrt(n)
  lower <- round(sampleMean - error, digits = 2)
  upper <- round(sampleMean + error, digits = 2)
  return(c(lower, upper))
}

Vec1 <- rnorm(100, mean = 10, sd = 2)
CI <- CI95(Vec1)
print(CI)
```


<span style="color:blue"> When we increase the number of coins for each
trial, the distribution becomes more narrow and peaked around the
expected proportion of tails. This is because, as the number of coins
increases, the law of large numbers states that the proportion of tails
will converge to the true probability of getting tails, which is 0.5 for
a fair coin. Therefore, with more coins, we have a higher chance of
observing the expected proportion of tails and less chance of observing
extreme proportions.</span>

<span style="color:blue"> For example, when we have 15 coins in each
trial, the distribution is more spread out and has a wider range of
proportion values compared to when we have 1500 coins in each trial.
Additionally, the distribution becomes more symmetric and bell-shaped as
the number of coins increases, indicating that the observations are more
likely to be centered around the expected proportion of tails. </span>
