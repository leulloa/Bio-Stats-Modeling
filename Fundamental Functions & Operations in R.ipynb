``` r
myQF <- function(a, b, c){
  disc <- b^2 - 4*a*c
  if (disc < 0){
    return(NULL)
  } else if (disc == 0){
    return (-b/(2*a))
  } else {
    root1 <- (-b + sqrt(disc))/(2*a)
    root2 <- (-b - sqrt(disc))/(2*a)
    return(c(root1, root2))
  }
}
```
